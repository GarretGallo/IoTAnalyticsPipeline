### Data Cleansing
- Here is now that we found our data - this is where some basic data cleansing is going to be performed
- Specifically here:
    * We are going to be creating two separate tables: 1: `H2S and Condition` & 2: `Monitor Maintenance`
- This will then be pushed to our bronze schema

**Importing Data and Functions**

```python
from pyspark.sql.functions import col
```

```python
bronze_df = spark.read.table("gasmonitordata.bronze.ingesteddata")
```

```python
print(bronze_df.count())
```

**Creating Table 1: H2S and Conditions**

```python
h2sandconditions = bronze_df. \
    withColumn('timeStamp', col('timestamp').cast('timestamp')). \
    select('monitorNum',
           'h2sReading',
           'humidity',
           'pressure',
           col('temperatureMH').alias('tempMH'),
           col('timeStamp'))

h2sandconditions.write.mode('overwrite').format('delta').saveAsTable('gasmonitordata.silver.monitorh2sandconditions')
```

**Creating Table 2: Monitor Maintenance**

```python
maintenance = bronze_df. \
    withColumn('timeStamp', col('timestamp').cast('timestamp')). \
    select('monitorNum',
           'batteryHealth',
           'daysUntilCalibration',
           'technicalFailure',
           'timeStamp')
                               
maintenance.write.mode('overwrite').format('delta').saveAsTable('gasmonitordata.silver.monitormaintenance')
```

```python
print(maintenance.count())
```
